# Data Generation Part

```python
# Importing and Stuff
import numpy as np
import pandas as pd
import random as r
from datetime import datetime
# Global Variables
rows_count = 500
np.random.seed(1)
```

```python
def create_unique_array(min, max, size):
    arr = {min}
    while len(arr) < size:
        n = np.random.randint(
            min,
            max
        )
        arr.add(n)
    arr = list(arr)
    return arr
def create_dates(size):
    dates = []
    today = datetime.today().strftime('%Y-%m-%d')
    month = int(today[5:7])
    year = int(today[:4])
    for i in range(size):
        month_rand = np.random.randint(month - 6, month)
        day_rand = np.random.randint(1, 28)
        string = f"{year}-{str(month_rand).zfill(2)}-{str(day_rand).zfill(2)}"
        dates.append(string)
    return dates
```

```python
# Table Schema
data = {
    "CustomerID": np.array(create_unique_array(1000, 9999,rows_count)),
    "Age": np.random.randint(18, 70, size=rows_count),
    "Gender": np.random.choice(['Male', 'Female', 'Other'], rows_count),
    "ProductID": np.array(create_unique_array(100, 999, rows_count)),
    "PurchaseAmount": np.random.uniform(50, 500, rows_count),
    "Rating": np.random.randint(1, 6, size=rows_count),
    "Review": np.random.choice(["Good", "Average", "Excellent", "Poor", "Satisfactory"], rows_count),
    "PurchaseDate": np.array(create_dates(rows_count))
}
```

```python
# Save to CSV
data_df = pd.DataFrame(data)
print(data_df)
data_df.to_csv('customer_feedback.csv')
```

# Data Processing Part

```python
data_df = pd.read_csv('customer_feedback.csv')
data_df.head(10)
```

```python
data_df.info()
data_df.describe()
```

```python
def introduce_missing_values(column, percentage):
    rows = len(column)
    rows_num_missing = int(rows*(percentage/100))
    missing_indicies = np.random.choice(rows, rows_num_missing)
    if column.dtype == 'float64' or column.dtype == 'int64':
        column[missing_indicies] = np.nan
    else:
        column[missing_indicies] = None
    return column
```

```python
#Introduce missing values
data_df['Age'] = introduce_missing_values(data_df['Age'], 32)
data_df['PurchaseAmount'] = introduce_missing_values(data_df['PurchaseAmount'], 40)
data_df['Rating'] = introduce_missing_values(data_df['Rating'], 20)
data_df.isnull().sum()
```

```python
# Missing value Handling
data_df['Age'] = data_df['Age'].fillna(data_df['Age'].mean())
data_df['PurchaseAmount'] = data_df['PurchaseAmount'].fillna(data_df['PurchaseAmount'].median())
data_df['Rating'] = data_df['Rating'].fillna(data_df['Rating'].mode())
data_df.isnull().sum()
```

```python
# Duplicate rows Deletion
data_df = data_df.drop_duplicates()
```
